```python
import pandas as pd
from matplotlib import pyplot
from pandas.plotting import lag_plot
```

```python
series = pd.read_csv("E:/datasets/daily_min_temperatures/daily-min-temperatures.csv", header=0, index_col=0, parse_dates=True,
squeeze=True)
print(type(series))
print(series.head())
```

# Basic features engineering

```python
df = pd.DataFrame()
df['month'] = [series.index[i].month for i in range(len(series))]
df['day'] = [series.index[i].day for i in range(len(series))]
df['temperature'] = [series[i] for i in range(len(series))]

df.head()
```

```python
temps = pd.DataFrame(series.values)
df = pd.concat([temps.shift(1), temps], axis=1)
df.columns = ['t', 't+1']
df.head()
```

```python
# lag features of 3

df = pd.concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
df.columns = ['t-2', 't-1', 't', 't+1'] 
df.head()
```

# Rolling Window Statistics

```python
shift = temps.shift(1)
window = shift.rolling(window=2)
means = window.mean()
df = pd.concat([means, temps], axis=1)
df.columns = ['mean(t-1, t)', 't+1']
df.head()
```

```python
window_width = 3
shift = temps.shift(window_width - 1)
window = shift.rolling(window=window_width)

df = pd.concat([window.min(), window.mean(), window.max(), temps], axis=1)
df.columns = ['min', 'mean', 'max', 't+1']
df.head()
```

# Expanding Window Statistics

```python
window = temps.expanding()
df = pd.concat([window.min(), window.mean(), window.max(), temps.shift(-1)], axis=1)
df.columns = ['min', 'mean', 'max', 't+1']
df.head()
```

```python
groups = series.groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()

for name, group in groups:
    years[name.year] = group.values

years.boxplot()
pyplot.show()
```

```python
one_year = series['1990']
groups = one_year.groupby(pd.Grouper(freq='M'))
months = pd.concat([pd.DataFrame(x[1].values) for x in groups], axis=1)
#months = pd.DataFrame(months)
months.columns = range(1, 13)
months.boxplot()
pyplot.show()
```

```python
groups = series.groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()
for name, group in groups:
    years[name.year] = group.values
years = years.T
pyplot.matshow(years, interpolation=None, aspect='auto', cmap='inferno')
pyplot.show()
```

```python
# Compare months within a year, col=month, row=day
one_year = series['1990']
groups = one_year.groupby(pd.Grouper(freq='M'))
months = pd.concat([pd.DataFrame(x[1].values) for x in groups], axis=1)
months.columns = range(1, 13)

pyplot.matshow(months, interpolation=None, aspect='auto', cmap='inferno')
pyplot.show()
```

```python
values = pd.DataFrame(series.values)
lags = 7
columns = [values]
for i in range(1, (lags + 1)):
    columns.append(values.shift(i))
df = pd.concat(columns, axis=1)
columns = ['t']

for i in range(1, (lags + 1)):
    columns.append('t-' + str(i))
df.columns = columns
pyplot.figure(1)

for i in range(1, (lags + 1)):
    ax = pyplot.subplot(240 + i)
    ax.set_title('t vs t-' + str(i))
    pyplot.scatter(x=df['t'].values, y=df['t-'+str(i)].values)
pyplot.show()
```
