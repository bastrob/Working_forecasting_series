# Basic features engineering

```python
df = pd.DataFrame()
df['month'] = [series.index[i].month for i in range(len(series))]
df['day'] = [series.index[i].day for i in range(len(series))]
df['temperature'] = [series[i] for i in range(len(series))]

df.head()
```

# Lag features

```python
temps = pd.DataFrame(series.values)
df = pd.concat([temps.shift(1), temps], axis=1)
df.columns = ['t', 't+1']
df.head()
```

```python
# lag features of 3

df = pd.concat([temps.shift(3), temps.shift(2), temps.shift(1), temps], axis=1)
df.columns = ['t-2', 't-1', 't', 't+1'] 
df.head()
```

# Rolling Window Statistics

```python
shift = temps.shift(1)
window = shift.rolling(window=2)
means = window.mean()
df = pd.concat([means, temps], axis=1)
df.columns = ['mean(t-1, t)', 't+1']
df.head()
```

```python
window_width = 3
shift = temps.shift(window_width - 1)
window = shift.rolling(window=window_width)

df = pd.concat([window.min(), window.mean(), window.max(), temps], axis=1)
df.columns = ['min', 'mean', 'max', 't+1']
df.head()
```

# Expanding Window Statistics

```python
window = temps.expanding()
df = pd.concat([window.min(), window.mean(), window.max(), temps.shift(-1)], axis=1)
df.columns = ['min', 'mean', 'max', 't+1']
df.head()
```

## Detrend by model fitting

```python
# use a linear model to detrend a time series
from pandas import read_csv
from pandas import datetime
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
import numpy

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')
series = read_csv('E:/datasets/shampoo.csv', header=0, index_col=0, parse_dates=True,
    squeeze=True, date_parser=parser)

# fit linear model
X = [i for i in range(0, len(series))]
X = numpy.reshape(X, (len(X), 1))
y = series.values
model = LinearRegression()
model.fit(X, y)

# calculate trend
trend = model.predict(X)

# plot trend
pyplot.plot(y)
pyplot.plot(trend)
pyplot.show()

# detrend
detrended = [y[i]-trend[i] for i in range(0, len(series))]

# plot detrended
pyplot.plot(detrended)
pyplot.show()
```

```python
# model seasonality with a polynomial model
from pandas import read_csv
from matplotlib import pyplot
from numpy import polyfit
series = read_csv('E:/datasets/daily_min_temperatures/daily-min-temperatures.csv', header=0, index_col=0,
parse_dates=True, squeeze=True)
# fit polynomial: x^2*b1 + x*b2 + ... + bn
X = [i%365 for i in range(0, len(series))]
y = series.values
degree = 4
coef = polyfit(X, y, degree)
print('Coefficients: %s' % coef)

# create curve
curve = list()
for i in range(len(X)):
    value = coef[-1]
    for d in range(degree):
        value += X[i]**(degree - d) * coef[d]
    curve.append(value)

# plot
pyplot.plot(y)
pyplot.plot(curve, color='red', linewidth=3)
pyplot.show()

# create seasonality adjusted
values = series.values
diff = list()

for i in range(len(values)):
    value = values[i] - curve[i]
    diff.append(value)

pyplot.plot(diff)
pyplot.show()
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

# We want to predict the last 7 days given all prior observations
# Simplest model -> persist the last observation and do 7 1day forecasts
values = DataFrame(series.values)
dataframe = concat([values.shift(1), values], axis=1)
dataframe.columns = ['t', 't+1']

X = dataframe.values
train_len = len(X) - 7

# Drop first NaN row
train, test = X[1:train_len], X[train_len:]
X_train, y_train = train[:, 0], train[:, 1]
X_test, y_test = test[:, 0], test[:, 1]

predictions_naive = list()

def persistence_model(x):
    return x

for x in X_test:
    yhat = persistence_model(x)
    predictions_naive.append(yhat)

rmse_naive = sqrt(mean_squared_error(predictions_naive, y_test))
print('RMSE naive model: ', rmse_naive)
pyplot.plot(y_test)
pyplot.plot(predictions_naive, color='red')
pyplot.show()
```

```python
series = read_csv('E:/datasets/daily_min_temperatures/daily-min-temperatures.csv', header=0, index_col=0,
    parse_dates=True, squeeze=True)
# split dataset
X = series.values
train, test = X[1:len(X)-7], X[len(X)-7:]
# train autoregression
model = AR(train)
model_fit = model.fit()
window = model_fit.k_ar
coef = model_fit.params
# walk forward over time steps in test
history = train[len(train)-window:]
history = [history[i] for i in range(len(history))]
predictions = list()

for t in range(len(test)):
    lenght = len(history)
    lag = [history[i] for i in range(lenght-window, lenght)]
    yhat = coef[0]
    for d in range(window):
        yhat += coef[d+1] * lag[window-d-1]
    
    obs = test[t]
    history.append(obs)
    predictions.append(yhat)
    print('predicted=%f, expected=%f' % (yhat, obs))

rmse = sqrt(mean_squared_error(test, predictions))
print('Test RMSE: %.3f' % rmse)
# plot
pyplot.plot(test)
pyplot.plot(predictions, color='red')
pyplot.show()
```

## Autoregression of Residual Error

```python
# Building on the persistence model defined before

series = read_csv('E:/datasets/daily-total-female-births.csv', header=0, index_col=0,
    parse_dates=True, squeeze=True)

values = DataFrame(series.values)
dataframe = concat([values.shift(1), values], axis=1)
dataframe.columns = ['t', 't+1']

X = dataframe.values
train_size = int(len(X) * 0.66)
train, test = X[1:train_size], X[train_size:]
X_train, y_train = train[:, 0], train[:, 1]
X_test, y_test = test[:, 0], test[:, 1]

# persistence model on the training set
train_predictions = [x for x in X_train]

# calculate residuals
train_residuals = [y_train[i] - train_predictions[i] for i in range(len(train_predictions))]

# error
rmse = sqrt(mean_squared_error(y_train, train_predictions))
print('Train RMSE: %.3f' % rmse)

# model the training set residuals
model = AR(train_residuals)
model_fit = model.fit()
window = model_fit.k_ar
coef = model_fit.params
print('Lag=%d, Coef=%s' % (window, coef))


# walk forward over time steps in test
history = train_residuals[len(train_residuals) - window:]
history = [history[i] for i in range(len(history))]

predictions_error = list()
expected_error = list()


for i in range(len(y_test)):
    yhat = X_test[i]
    error = y_test[i] - yhat
    expected_error.append(error)
    
    # predict error
    lenght = len(history)
    lag_error = [history[i] for i in range(lenght-window, lenght)]

    pred_error = coef[0]
    for d in range(window):
        pred_error += coef[d+1] * lag_error[window-d-1]
        
    predictions_error.append(pred_error)
    history.append(error)
    print('predicted error=%f, expected error=%f' % (pred_error, error))
    
# plot predicted error
pyplot.plot(expected_error)
pyplot.plot(predictions_error, color='red')
pyplot.show()
```

```python
# Building on the persistence model defined before

series = read_csv('E:/datasets/daily-total-female-births.csv', header=0, index_col=0,
    parse_dates=True, squeeze=True)

values = DataFrame(series.values)
dataframe = concat([values.shift(1), values], axis=1)
dataframe.columns = ['t', 't+1']

X = dataframe.values
train_size = int(len(X) * 0.66)
train, test = X[1:train_size], X[train_size:]
X_train, y_train = train[:, 0], train[:, 1]
X_test, y_test = test[:, 0], test[:, 1]

# persistence model on the training set
train_predictions = [x for x in X_train]

# calculate residuals
train_residuals = [y_train[i] - train_predictions[i] for i in range(len(train_predictions))]

# model the training set residuals
model = AR(train_residuals)
model_fit = model.fit()
window = model_fit.k_ar
coef = model_fit.params
print('Lag=%d, Coef=%s' % (window, coef))


# walk forward over time steps in test
history = train_residuals[len(train_residuals) - window:]
history = [history[i] for i in range(len(history))]

predictions = list()

for i in range(len(y_test)):
    yhat = X_test[i]
    error = y_test[i] - yhat
    
    # predict error
    lenght = len(history)
    lag_error = [history[i] for i in range(lenght-window, lenght)]
    pred_error = coef[0]
    
    for d in range(window):
        pred_error += coef[d+1] * lag_error[window-d-1]
    
    # correct the prediction
    yhat = yhat + pred_error
    predictions.append(yhat)
    history.append(error)
    print('predicted =%f, expected =%f' % (yhat, y_test[i]))

# error
rmse = sqrt(mean_squared_error(y_test, predictions))
print('Test RMSE: %.3f' % rmse)

# plot predicted error
pyplot.plot(y_test)
pyplot.plot(predictions, color='red')
pyplot.show()
```
